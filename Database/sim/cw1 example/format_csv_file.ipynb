```python
import pandas as pd

# load a csv file to Pandas' dataframe for initial exploration and formating
df = pd.read_csv('sf_crime_data.csv')
df.head()
```

```python
# remove ',' from the column values, 'Category' and 'Resolution' in particular, and replace it with '/' character
# if a column value contains ',', then MySQL will fail to load the data.
# MySQL cannot distinguish between ',' contained in the column and the one used to separate the columns.  
df['Resolution'] = df['Resolution'].apply(lambda x: x.replace(',', ' /'))
df['Descript'] = df['Descript'].apply(lambda x: x.replace(',', ''))
df['street'] = df['Address'].apply(lambda x: x.split(' / ')[-1])
df['block'] = df['Address'].apply(lambda x: x.split(' / ')[0].split(' Block of ')[0])

df.columns = ['timestamp', 'category', 'description', 'DayOfWeek', 'district', 'resolution', 'Address', 'latitude', 'longitude', 'street', 'block']
df['category'] = df['category'].apply(lambda x: x.replace('/', ' / '))
df = df.drop(columns=['DayOfWeek', 'Address'])
# add index column


# preview re-formatted data
df.head()
```

```python
# export formatted data to a csv file, sf-crime-data.csv 
df.to_csv('sql-loadable-clf-crime-data.csv', index=False)

# create a subset of data, e.g. 1st 1,000 rows, for development and testing of SQL codes 
df.iloc[:1000].to_csv('tiny-clf-crime-data.csv', index=False)
```
